# NOT: Bir bilim insanının doğduğu ülke,kültürel ve sosyal faktörler nedeniyle bilimsel başarıları üzerinde etkili olabilir. Bu nedenle analizimde birth_country sütununu kullanacağım. Yapmış olduğum bu analiz demografik bir analiz olacaktır.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('python_proje.csv')
df
```

```python
df.columns
```

```python
# Sütunların veri tipini kontrol edelim.
df.info()
```

```python
# Tarihleri DateTime olarak değiştirelim

df['birth_date'] = pd.to_datetime(df['birth_date'], errors='coerce')
df['death_date'] = pd.to_datetime(df['death_date'], errors='coerce')
df.info()
```

```python
# prize_share dönüşümü ( "1/..." formatında olduğu için floata dönüştürelim)

df['prize_share'] = df['prize_share'].apply(
    lambda x: eval(x) if isinstance(x, str) and '/' in x else x
).astype(float)
df.head()
```

```python
# Veri setinde tekrarlayan satırlar olup olmadığını kontrol edelim.
df[df.duplicated()]
```

```python
# Eksik veri var mı kontrol edelim.
print(df.isnull().sum())
```

```python
# Eksik verileri heatmap ile görselleştirelim

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Eksik verileri dolduralım

df['motivation'].fillna('Unknown', inplace=True)
df['birth_city'].fillna('Unknown', inplace=True)
df['birth_country'].fillna('Unknown', inplace=True)
df['sex'].fillna('Unknown', inplace=True)
df['organization_name'].fillna('Unknown', inplace=True)
df['organization_city'].fillna('Unknown', inplace=True)
df['organization_country'].fillna('Unknown', inplace=True)

df.head()
```

# 1.Nobel Ödüllerini en çok kazanan ilk on ülkeyi bulunuz.

```python
# 'Unknown' değeri olan ülkeleri filtreleyerek gerçek ülkeleri alalım.
valid_countries = df[df['birth_country'] != 'Unknown']

# 'birth_country' sütunundaki 'Unknown' değeri olmayan ülkelerin sayısını hesaplayalım.
country_counts = valid_countries['birth_country'].value_counts()

# En çok ödül alan ilk on ülkeyi seçelim.
top_10_countries = country_counts.head(10)

print(top_10_countries)
```

# 2.Nobel Ödüllerini ilk kazanan kadınlar listeleyiniz.

```python
# Sadece kadın olanları filtreleyip ilk 10 kadını bulalım.
female_laureates = df[df['sex'] == 'Female']
female_laureates.head(10)[['year','category','full_name']]
```

# 3. Nobel Ödüllerini kazanan ilk erkekleri listeleyiniz.

```python
# Sadece erkek olanları filtreleyip ilk 10 erkeği bulalım.
male_laureates = df[df['sex'] == 'Male']
male_laureates.head(10)[['year','category','full_name']]
```

# 4. Nobel ödülünü en çok kazanan ülkenin hangi yıldan itibaren hakimiyet sağladığını görselleştirip bu hakimiyette rol oynayan şeyler nelerdir? İçgörülerinizi paylaşır mısınız?

```python
# 'birth_country' sütunundaki ödül kazananların ülkelerini gruplayarak toplam sayılarını hesaplayalım.
country_counts = df['birth_country'].value_counts()

# En çok ödül kazanan ilk 20 ülkeyi seçelim.
top_countries = country_counts.head(10)

# Ülkelerin kazanma sayılarını ve isimlerini alalım.
countries = top_countries.index
counts = top_countries.values
```

```python
# Ülkelerin kazanma sayılarını görselleştirelim.
plt.figure(figsize=(5, 5))
plt.bar(countries, counts, color='skyblue')
plt.xlabel('Ülkeler')
plt.ylabel('country_counts')
plt.title('Nobel Ödüllerini En Çok Kazanan Ülkeler')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Her yıl başına ABD doğumlu kazananların oranlarını görüntüleyelim.

df['usa_born_winner'] = df.birth_country == 'United States of America'
prop_usa_winners = df.groupby('year', as_index=False)['usa_born_winner'].mean()
prop_usa_winners.head(35)
```

```python
# Yukarıdaki tabloyu görselleştirelim.

plt.figure(figsize=(12, 6))
plt.plot(prop_usa_winners['year'], prop_usa_winners['usa_born_winner'], marker='o', color='b')
plt.xlabel('Year')
plt.ylabel('Proportion of USA-born Winners')
plt.title('Proportion of USA-born Nobel Prize Winners over Time')
plt.grid(True)
plt.show()
```

Şimdi yukarıdaki tabloyu diğer ülkelerle kıyaslamak için birleştirelim.

```python
# Veri setinden ülkeleri ve yılları gruplayarak toplam ödül sayılarını hesaplayalım.
country_year_counts = df.groupby(['birth_country', 'year']).size().unstack(fill_value=0)

# 'birth_country' sütunundaki en çok ödül kazanan ilk 10 ülkeyi seçelim.
top_countries = country_counts.head(10).index

# Seçilen ülkelerin yıllara göre ödül sayılarını görselleştirelim.
plt.figure(figsize=(12, 8))
for country in top_countries:
    plt.plot(country_year_counts.loc[country], label=country)

plt.title('Yıllara Göre En Çok Ödül Kazanan Ülkelerin Değişimi')
plt.xlabel('Yıl')
plt.ylabel('Ödül Sayısı')
plt.legend()
plt.grid(True)
plt.show()
```

# 5.Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.

```python
# Veri setindeki kişilerin yaşını hesaplayalım ve bunu yeni bir sütuna ekleyelim.

df['age_difference'] = df['year'] - df['birth_date'].dt.year
```

```python
# Gerekli sütunları seçerek yeni bir veri çerçevesi oluşturalım.
new_data = df[['sex', 'age_difference', 'category', 'year']]
# Veriyi yazdıralım.
new_data
```

```python
#Görselleştirme için ilgili sütunları seçelim.
data = df[['year', 'category', 'sex', 'age_difference']]

# age_difference sütunundaki eksik değerleri filtreleyelim.
data = df.dropna(subset=['age_difference'])

# Her ödül kategorisi için cinsiyete dayalı olarak yıl ve yaş arasındaki ilişkiyi görselleştirmek amacıyla bir dağılım grafiği oluşturalım.
categories = data['category'].unique()
colors = {'M': 'blue', 'F': 'red'}

for category in categories:
    plt.figure(figsize=(8, 5))
    plt.title(f'Nobel Ödülü Sahiplerinin Cinsiyete Göre Yıllara Göre Yaşı for {category} Category')
    
    for gender in ['Male', 'Female']:
        plt.scatter(data[(data['category'] == category) & (data['sex'] == gender)]['year'], 
                    data[(data['category'] == category) & (data['sex'] == gender)]['age_difference'], 
                     label=gender)
    
    plt.xlabel('Year')
    plt.ylabel('Age Difference')
    plt.legend()
    plt.show()
```

# 6. 1938-1945 yılı arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.

```python
# 1938-1945 yılları arasındaki verileri filtreleyelim.
filtered_df = df[(df['year'] >= 1938) & (df['year'] <= 1945)]

# Kategorilere göre dağılım
category_counts = filtered_df['category'].value_counts()

plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Kategori')
plt.ylabel('Kazanan Sayısı')
plt.title('1938-1945 Nobel Ödülü Kategorileri Dağılımı')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Ülkelere göre dağılım
country_counts = filtered_df['birth_country'].value_counts()

plt.figure(figsize=(12, 6))
country_counts.plot(kind='bar', color='lightgreen')
plt.xlabel('Ülke')
plt.ylabel('Kazanan Sayısı')
plt.title('1938-1945 Nobel Ödülü Kazananların Ülkelere Göre Dağılımı')
plt.grid(axis='y')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show
```

# 7. 1947-1991 yılları arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.(Her kategori için ayrı bir grafik olması istenmektedir)

```python
df_subset = df[(df['year'] >= 1947) & (df['year'] <= 1991)]

# Her kategori için ayrı grafikler oluşturalım
categories = df_subset['category'].unique()
plt.figure(figsize=(12, 8))

for i, category in enumerate(categories, 1):
    plt.subplot(2, 3, i)
    category_data = df_subset[df_subset['category'] == category]
    country_counts = category_data['birth_country'].value_counts().head(5)  # Sadece en çok ödül alan 5 ülkeyi gösterelim
    country_counts.plot(kind='bar', color='skyblue')
    plt.title(f'{category} Ödülleri')
    plt.xlabel('Doğum Ülkesi')
    plt.ylabel('Kazanan Sayısı')

plt.tight_layout()
plt.show()
```

# 8. Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin.(Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.


```python

# Belirli kategorilerdeki kişileri filtreleyin (Kimya, Edebiyat, Barış, Fizik, Tıp)
categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Medicine']
filtered_df = df[df['category'].isin(categories)]

# 2000 sonrasındaki kişileri filtreleyelim.
filtered_df = filtered_df[filtered_df['year'] > 2000]
filtered_df['age_difference'] = filtered_df['year'] - filtered_df['birth_date'].dt.year
filtered_df[['year','full_name','birth_country','category','birth_date','age_difference']]
```

```python
plt.figure(figsize=(15,10))

sns.scatterplot(
    data=filtered_df, 
    x='birth_country', 
    y='age_difference', 
    s=100, 
    color='blue'
)

plt.title("2000 Sonrası Nobel Ödülü Kazananların Ülkeleri ve Yaşları")
plt.xlabel('Doğum Ülkesi')
plt.ylabel('Yaş')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
# Veri setindeki sadece 2000 yılından sonraki girişleri filtreleyelim.
data_after_2000 = df[df['year'] > 2000]

# Kimya kategorisindeki kişilerin ülkelerini görselleştirelim.
plt.figure(figsize=(8,4))
data_chemistry = data_after_2000[data_after_2000['category'] == 'Chemistry']
plt.scatter(data_chemistry['birth_country'], data_chemistry['age_difference'], color='b')
plt.xlabel('Ülkeler')
plt.ylabel('Yaşlar')
plt.title('Kimya Kategorisindeki Nobel Ödüllü Kişilerin Ülkelerine Göre Yaş Dağılımı')
plt.xticks(rotation=90)
plt.show()
```
